```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

data1_copy = data1.copy()

numeric_features = data1_copy.select_dtypes(include='number')
# Normalizing the data using StandardScaler()
scaler = StandardScaler()
df_normalized = scaler.fit_transform(numeric_features)
data1_copy[numeric_features.columns] = df_normalized
```

```python
data1.describe()
```

```python
# Creating a dataset with lesser outliers
threshold = 3
outlier_mask = (np.abs(data1_copy.select_dtypes(include='number')) > threshold).any(axis=1)
data1_less_outliers = data1_copy.select_dtypes(include='number')[~outlier_mask]
data1=data1_less_outliers
```

```python
# A Correlation Matrix for the dataset with lesser outliers
correlation_matrix=data1.corr()
#data1['duration_x_energy']=data1['duration_ms']*data1['energy']
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Our dataset did not contain any classes, hence, we added a feature with 2 classes.
# The feature 'popular' is based on 'track_popularity'.
# If 'track_popularity' is greater than 0.6, then 'popular' value equals 1, else 0.
data1['popular']=data1['track_popularity'].apply(lambda x:1 if x>0.6 else 0)
```

```python
data1['popular'].value_counts()
```

```python
# Combined 2 features to check which are most correlating
data1['duration_+_acousticness'] = data1['duration_ms'] + data1['acousticness'] 
data1['valence_+_energy'] = data1['valence'] + data1['energy'] 
data1['acousticness_+_energy'] = data1['acousticness'] + data1['valence_+_energy'] 
data1['acousticness_+_duration_ms'] = data1['acousticness'] - data1['duration_ms'] 

# Combined 4 features (just for experimenting)
data1['duration_+_energy_+_valence_+_energy'] = data1['duration_+_acousticness'] + data1['valence_+_energy'] 

# Combined 5 features
data1['du_+_va_+_en_+_ac_+_da'] = data1['duration_ms'] +  data1['valence'] + data1['energy'] +  data1['acousticness'] + data1['danceability'] 

# Combined 10 features
data1['10xFeatures'] = data1['duration_ms'] +  data1['valence'] + data1['energy'] +  data1['acousticness'] + data1['danceability'] + data1['artist_popularity'] + data1['key'] +data1['loudness']+ data1['liveness']+ data1['instrumentalness']

# Correlation matrix
correlation_matrix = data1.corr()
plt.figure(figsize=(14, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Dropped unnecessary columns
data1.drop(columns=['tempo','time_signature','mode'],inplace=True)
print(data1.columns)
correlation_matrix = data1.corr()
plt.figure(figsize=(14, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
"""
Feature selection is best first done by correlation matrices as some features are not important
Non numeric features have been dropped but dataset can be farther reduced
If certain features remain important but a reduced number of features is desired
PCA is used to reduce the number of features and preserves variance 
"""
from sklearn.decomposition import PCA

data1_PCA=data1.drop(columns=['track_popularity'])
Y = data1['track_popularity'] 
data1_numeric = data1_PCA.select_dtypes(include='number')  
# Reduce to 2 features
num_components = 2 
pca = PCA(num_components)
data1_reduced2 = pca.fit_transform(data1_numeric)
# Reduce to 5 features
num_components = 5 
pca = PCA(num_components)
data1_reduced5 = pca.fit_transform(data1_numeric)
# Reduce to 10 features
num_components = 10
pca = PCA(num_components)
data1_reduced10 = pca.fit_transform(data1_numeric)

result2 = pd.concat([Y, pd.DataFrame(data1_reduced2,columns=['PCA1','PCA2'])], axis=1)
print(result2)
result5 = pd.concat([Y, pd.DataFrame(data1_reduced5,columns=['PCA1','PCA2', 'PCA3', 'PCA4', 'PCA5'])], axis=1)
print(result5)
result10 = pd.concat([Y, pd.DataFrame(data1_reduced10,columns=['PCA1','PCA2', 'PCA3', 'PCA4', 'PCA5', 'PCA6','PCA7', 'PCA8', 'PCA9', 'PCA10'])], axis=1)
print(result10)
```
